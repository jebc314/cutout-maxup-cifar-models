```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# ResNet-110 CIFAR-10 Comparison

```python
with open("cifar10-cutout/resnet110.txt") as f:
    r10c_percents = []
    for line in f.readlines():
        if "*" in line:
            percentIndex = line.index("%")
            r10c_percents.append(float(line[8:percentIndex]))

with open("cifar10-maxup-cutout/resnet110.txt") as f:
    r10mc_percents = []
    for line in f.readlines():
        if "*" in line:
            percentIndex = line.index("%")
            r10mc_percents.append(float(line[8:percentIndex]))

print (f"ResNet-110 CIFAR10 with cutout {max(r10c_percents)} with maxup-cutout {max(r10mc_percents)}")
plt.plot(r10c_percents, label="Cutout")
plt.plot(r10mc_percents, label="Cutout & MaxUp")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy (Percentage)")
plt.title("ResNet-110 on CIFAR-10")
```

# ResNet-110 CIFAR-100 M Comparison

```python
resnet_m = {
    1:[],
    4:[],
    10:[]
}

for m in resnet_m:
    with open(f"resnet-110-ms/m-{m}.txt") as f:
        for line in f.readlines():
            if "*" in line:
                percentIndex = line.index("%")
                resnet_m[m].append(float(line[8:percentIndex]))

print ("ResNet-110 CIFAR100 with maxup-cutout")
for m in resnet_m:
    print(f"m of {m}: {max(resnet_m[m])}")
    plt.plot(resnet_m[m], label=f"m={m}")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy (Percentage)")
plt.title("ResNet-110 on CIFAR-100 with varying m's")
```
