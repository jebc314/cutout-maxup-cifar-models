```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# CIFAR-10

```python
with open("cifar10-cutout/resnet110.txt") as f:
    percents = []
    for line in f.readlines():
        if "*" in line:
            percentIndex = line.index("%")
            percents.append(float(line[8:percentIndex]))

plt.plot(percents)
print(max(percents))
```
